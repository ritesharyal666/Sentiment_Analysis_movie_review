```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,SimpleRNN
from tensorflow.keras.preprocessing import sequence
```

```python
from tensorflow.keras.datasets import imdb
(x_train,y_train),(x_test,y_test)= imdb.load_data()
```

```python
max_features=100000
```

```python
word_index=imdb.get_word_index()
```

```python
indexed_word ={value:key for key,value in word_index.items()}
```

```python
sample_review=' '.join(indexed_word.get(i-3,'?') for i in x_train[0])
```

```python
sample_review
```

```python
max_length_of_sentences=500
```

```python
x_train=sequence.pad_sequences(x_train,max_length_of_sentences)
```

```python
x_test=sequence.pad_sequences(x_test,max_length_of_sentences)
```

```python
dimension=128
```

```python
model = Sequential()
model.add(Embedding(max_features,dimension))#EmbeddingLayer
model.add(SimpleRNN(128,activation='relu'))#128 neurons
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es=EarlyStopping(monitor='val_loss',patience=4,restore_best_weights=True)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(
    x_train,y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    callbacks=[es],
)
```

```python
def decoded_review(encoded_review):
  return ' '.join(indexed_word.get(i-3,'?')for i in encoded_review)
```

```python
def preprocess_text(text):
  words =text.lower().split()
  encoded_review=[word_index.get(word,2)+3 for word in words]
  padded_review=sequence.pad_sequences([encoded_review],maxlen=500)
  return padded_review
```

```python
def predict_sentiment(review):
  preprocessed_input = preprocess_text(review)
  prediction = model.predict(preprocessed_input)
  sentiment = 'Positive' if prediction[0][0]>0.5 else 'Negative'
  return sentiment,prediction[0][0]
```

```python
example_review='The movie was fantastic, acting was great ,plot good'
sentiment,prediction_score=predict_sentiment(example_review)
```
